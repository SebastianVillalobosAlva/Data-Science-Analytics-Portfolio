````python
def melt_plotter(data,sort_var,id_var,var_name,
                 value_name,cols,num_values,ascend=False,sort_bool=True):
    '''
    Plot barplots of various columns in the same figure, using seaborn barplot
    
    Parameters
    -----------
    data :  Dataframe
    
    sort_var : string 
        Column name by which the dataframe will be sorted
        
    ascend : bool, default "False"
        Sort the dataframe in ascending o descending form
    
    id_var : tuple, list, or ndarray, optional
        Column(s) to use as identifier variables.
    
    var_name : scalar
        Name to use for the 'variable' column.
    
    value_name : scalar, default 'value'
        Name to use for the 'value' column.
    
    cols : string, list
        Column(s) to melt.
    
    num_values: int
        Number of examples to plot
    
    sort_bool : bool, default "True"
        Whether to sort or not the dataframe
    '''
    
    if sort_bool:
        data = data.sort_values(by=sort_var,ascending=ascend)[:num_values]

    data = pd.melt(data[cols], id_vars=id_var,var_name=var_name, value_name=value_name)
    plt.figure(figsize=(16,6))
    splot = sns.barplot(data=data, x=id_var, y=value_name,hue=var_name, palette='viridis')  
    for p in splot.patches:
        splot.annotate(format(p.get_height(), '.1f'), 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha = 'center', va = 'center', 
                       xytext = (0, 9), 
                       textcoords = 'offset points')
````

Based on the various plots made before, we'll try a clustering algorithm to see what groups are within our dataset. So we can narrow the number of countries in which CEO of NGO needs to look into.

````python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.base import BaseEstimator, TransformerMixin
````

First we will create our class in which we will store our labels.
<br>
After building our class, we are going to use `KMeans.inertia` to evaluate how many clusters and which scaler we should be using.

First we are going to prepare our data and use the elbow method to try and see how many clusters we can use.
<br>
We are going to drop the country column in our data, as our KMeans only needs numerical data.
After we have our data, we are going to scale it, this is where we try two types of scalers and see which one works best, we will try the `StandardScaler()` and `MinMaxScaler()`

````python
X = df.drop('country',axis=1)
#X
````

````python
StdScaler = StandardScaler()
MnMxScaler = MinMaxScaler()
````

````python
X_std = StdScaler.fit_transform(X)
X_mnmx = MnMxScaler.fit_transform(X)
````

After transforming our data, we will see which scaler works better. We'll be trying `X_std` first and then `X_mnmx`

````python
s_std = []
s_mnmx = []
list_k = list(range(1, 10))

for k in list_k:
    km = KMeans(n_clusters=k)
    km.fit(X_std)
    s_std.append(km.inertia_)
    km.fit(X_mnmx)
    s_mnmx.append(km.inertia_)

# Plot s_std against k
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,6))

ax1.plot(list_k, s_std, '-o')
ax1.set_xlabel(r'Number of clusters *k*')
ax1.set_ylabel('Sum of squared distance [StdScaler]');

ax2.plot(list_k, s_mnmx, '-o')
ax2.set_xlabel(r'Number of clusters *k*')
ax2.set_ylabel('Sum of squared distance [MnMxScaler]');
````

Now we know our `MinMaxScaler()` works better for this problem,but our data is not normally distributed, as seen in the distribution plots made before, so we will use our `StandardScaler()`, so we can normalize our data.

We are going to divide our data into 3 clusters

````python
class Clustering(BaseEstimator, TransformerMixin):
    
    def __init__(self,n_clusters):
        self.model = KMeans(n_clusters)
        self.scaler = StandardScaler() 
        
    def fit(self,X):
        X = self.scaler.fit_transform(X)
            
        if isinstance(X,pd.DataFrame):
            X = X.to_numpy()
            
        self.model.fit(X)
        
        return self.model.labels_
        
````

````python
Cluster = Clustering(3)
````

````python
Cluster.fit(X)
````

Now that we have our labels, we are going to put them in our dataframe and visualize our clusters

````python
df['label'] = Cluster.model.labels_
df
````

You can analyse the clusters by comparing how these three variables - [gdpp, child_mort and income] vary for each cluster of countries to recognise and differentiate the clusters of developed countries from the clusters of under-developed countries.
